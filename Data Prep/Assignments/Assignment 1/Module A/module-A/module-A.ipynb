__A4.__ (4 pts) Now collect the data from the `Character` column and load it into a list called `speakers` to complete the below function.

````python
# A4:Function(2/4)

def get_speakers(seinfeld):
    speakers = []
    
    # Need the first column of delimited data
    for line in seinfeld:
        # First, split out the columns by comma
        speaker = line[1]
        # Appending the speaker to the list
        speakers.append(speaker)
        speakers.sort()
    return speakers
````

__A5.__ _(2 points)_ Now complete the function to count the dialogue entries for the four major characters (`"JERRY"`, `"GEORGE"`, `"ELAINE"`, and `"KRAMER"`), using your output from the `get_speakers()` function and a dictionary (you are not allowed to use the `Counter` data structure for any component of this module). 

````python
# A5:Function(2/2)

def count_four_main(speakers):
    counts = {
        "JERRY" : 0,
        "GEORGE" : 0,
        "ELAINE" : 0,
        "KRAMER" : 0
    }

    speakers = get_speakers(seinfeld)

            
    for name in speakers:
        name_upper = name.upper().strip() # Convert the name to uppercase and strip white space

        # If the character speaks, the string should start and end with the name and only be alphabetical characters.
        if (name_upper.startswith("JERRY") and name_upper.endswith("JERRY")) and name_upper.isalpha():
            counts["JERRY"] += 1
        if (name_upper.startswith("GEORGE") and name_upper.endswith("GEORGE")) and name_upper.isalpha():
            counts["GEORGE"] += 1
        if (name_upper.startswith("ELAINE")  and name_upper.endswith("ELAINE")) and name_upper.isalpha():
            counts["ELAINE"] += 1
        if (name_upper.startswith("KRAMER") and name_upper.endswith("KRAMER")) and name_upper.isalpha():
            counts["KRAMER"] += 1
            
    return counts
````

__A6.__ _(3 points)_ Now complete the function to count the number of words spoken by each of the main characters using a dictionary. Note, this means returning to operate on the full `seinfeld` data object.

Note: For all word counting in this module, simply use the `.split()` method on strings.

````python
# A6:Function(3/3)

def count_words_spoken(seinfeld):
    words_spoken = {
        "JERRY" : 0,
        "GEORGE" : 0,
        "ELAINE" : 0,
        "KRAMER" : 0
    }
    
    for line in seinfeld:
    
        speaker = line[1].upper().strip()  # Extract and convert the speaker's name to uppercase
        dialogue = line[2]  # Extract the dialogue
        
        # Check if the speaker's name matches one of the characters
        if speaker in words_spoken:
            # Split the dialogue into words and add the word count to the character's total
            words_spoken[speaker] += len(dialogue.split())

    return words_spoken
````

__A8.__ _(6 points)_ Finally, complete the function below. This should load the `data/stop-words.txt` file into a list and then output a dictionary of the `N` most common words for each of the main characters that are _not_ on the this list of stop words. The output `most_common_words` should be a dictionary of `N`-element lists, with each element consisting of a `(word,count)` tuple. Note: each character's most-common words should be sorted from high-to-low.

````python
# A8:Function(6/6)

stop_words = [word.strip() for word in open("data/stop-words.txt", "r").readlines()]

def get_top_N(word_counts, stop_words, N=10):
    most_common_words = {
        "JERRY": [],
        "GEORGE": [],
        "ELAINE": [],
        "KRAMER": []
    }
    
    for speaker, words in word_counts.items():
        filtered_dict = {key: value for key, value in words.items() if key not in stop_words}
        sorted_dict = sorted(filtered_dict.items(), key=lambda item: item[1], reverse=True)
        most_common_words[speaker] = sorted_dict[:N]
    
    return most_common_words
````

````python
top_words=get_top_N(word_counts, stop_words)
top_words
````
