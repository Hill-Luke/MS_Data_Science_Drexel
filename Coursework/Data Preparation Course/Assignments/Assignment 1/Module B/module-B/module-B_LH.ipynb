````python
# B1:Function(5/5)

import csv

def load_metadata(path_to_file):
    reader = csv.reader(open(path_to_file, "r"))
    goodreads_books = list(reader)
    goodreads_books.pop(0)
    books_metadata = []
    
    for book in goodreads_books:
        
        if book[8]=='':
            book[8]=0

        authors=book[7]
        title=book[10]
        pub_year=int(float(book[8]))
        rating=float(book[12])
        rating_count=int(book[13])
        
        book_info = {
            'authors': authors,  # Author's information is in the 8th column (index 7)
            'original title': title,  # title is in the 11th column (index 10)
            'original publication year': pub_year,  # publication year is in the 9th column (index 8)
            'average rating': rating,  # average rating is in the 13th column (index 12)
            'ratings count': rating_count  # ratings count is in the 14th column (index 13)
        }
        books_metadata.append(book_info)
    
    return books_metadata

books_metadata = load_metadata("data/goodreads-books.csv")
books_metadata
````

__B3.__ _(5 points)_ Now complete the below function to take two arguments: the list, and an integer value for minimum ratings count. The new function should now sort _and_ filter the list, returning a list of books sorted by average rating that have been rated by more than a specified number of users. Here, you must use an approach using loops and `if` statements&mdash;in subsequent parts, comprehensions and the built-in `filter()` must be utilized instead (look up documentation and examples).

````python
# B3:Function(5/5)

def sort_and_filter(books_metadata, min_ratings_count):
    new_metadata = []
    
    for book in books_metadata:
        if book['ratings count'] > min_ratings_count:
            book_info = {
                'authors': book['authors'],
                'original title': book['original title'],
                'original publication year': (book['original publication year']),
                'average rating': (book['average rating']),
                'ratings count': (book['ratings count'])
            }
            new_metadata.append(book_info)    
    
    new_metadata = sorted(new_metadata, key=lambda x: x['average rating'], reverse=True)    
    return new_metadata
````
